```python
df = spark.table("gold.movimento_flat_diaria_06hs")

colunas_esperadas = [
    "nome_associado",
    "sobrenome_associado",
    "idade_associado",
    "vlr_transacao_movimento",
    "des_transacao_movimento",
    "data_movimento",
    "numero_cartao",
    "nome_impresso_cartao",
    "data_criacao_cartao",
    "tipo_conta",
    "data_criacao_conta"
]

assert df.columns == colunas_esperadas, "Schema da gold diferente do esperado"
```

```python
from pyspark.sql import functions as F

# Lista de colunas obrigatórias que não podem conter valores nulos
colunas_obrigatorias = [
    "vlr_transacao_movimento",
    "numero_cartao"
]

# Para cada coluna obrigatória, verifica se há valores nulos e lança erro se encontrar
for col in colunas_obrigatorias:
    qtd = df.filter(F.col(col).isNull()).count()
    assert qtd == 0, f"Encontrado nulo na coluna {col}"
```

```python
from pyspark.sql import functions as F

# Conta quantos cartões não estão no formato correto: 4 dígitos, 8 asteriscos, 4 dígitos
qtd_mascara_errada = df.filter(
    ~F.col("Numero_cartao").rlike(r"^\d{4}\*{8}\d{4}$")
).count()

# Garante que todos os cartões estão mascarados corretamente
assert qtd_mascara_errada == 0, "Máscara de cartão inválida"
```
